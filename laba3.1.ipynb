```python
import numpy as np
from matplotlib import pyplot as plt


x = np.linspace(-5, 5, 100)


y = x**2


plt.plot(x, y, '-')


plt.fill_between(x, y, 0, where=(y > 0), facecolor='g', alpha=0.6)


plt.title("Parabola Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


plt.show()
```
